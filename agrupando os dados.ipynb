```python
import pandas as pd 
dados = pd.read_excel('dados/IPCA MENSAL 2007-2019 limpa.xlsx')
dados.head()
```

```python
dados.columns
```

```python
dados.shape
```

```python
colunas_selec = dados.iloc[:,2:158]
```

```python
colunas_selec.head()
```

```python
col_data = colunas_selec.columns
lista_col = list()
for i in col_data:
    lista_col.append(i)
```

```python
lista_col
```

```python
data = dados.groupby(["Cadeia"])[lista_col].mean()
data
```

```python
data.to_csv('IPCA_DADOS_AGRUPADOS.csv',index=True, sep=';', encoding = 'Latin1')
```
